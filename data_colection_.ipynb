Download Price Data (TCS)

```python
import yfinance as yf
import pandas as pd

base_path = "/content/drive/MyDrive/stock-nlp-prediction"

ticker = "TCS.NS"

df = yf.download(ticker, start="2021-01-01", end="2025-01-01")

df.head()
```

Create lables

```python
# Flatten columns
df.columns = df.columns.get_level_values(0)

# Reset index
df = df.reset_index()

# Keep only needed columns
df = df[['Date', 'Close']]

# Create label
df['Next_Close'] = df['Close'].shift(-1)
df['Label'] = (df['Next_Close'] > df['Close']).astype(int)

df = df.dropna()

df.head()
```
